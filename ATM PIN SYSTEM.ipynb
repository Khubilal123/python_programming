# ATM LOGIN

```python
print("---configuration---")
CORRECT_PASSWORD = "1234"
MAX_ATTEMPTS = 4

#---ATM Login feature---
attempts_left=MAX_ATTEMPTS
login_succesful = False

while attempts_left>0:
    print(f"\nAttempts remaining:{attempts_left}")
    password = input("Enter password to access the converter:")
    if password == CORRECT_PASSWORD:
        print("Access granted! proceeding to Unit Converter.")
        login_successful = True
        break
    else:
        attempts_left -= 1
        if attempts_left>0:
            print("Incorrect password.please try again.")
if not login_successful:
       print(f"(\n*** ACCESS BLOCKED***")
       print(f"You have exceeded the maximum of {MAX_ATTEMPTS} incorrect attempts.")
```

# metric unit convertor

```python

 if login_succesful:
    print("\n--- Welcome to the Unit Converter ---")
    conversion_loop = True
    
    while conversion_loop:
        print("\nUnits: m, cm, dm, km")
        print("Enter 'Exit' to quit.")
        
        unit = input("Enter the unit: ")
        if unit.lower() == 'exit':
            print("Exiting converter")
            conversion_loop = False
            continue
        
        value = input(f"Enter value in {unit}: ")
        if value.lower() == 'exit':
            print("Exiting converter")
            conversion_loop = False
            continue
        
        try:
            value = float(value)  
        except ValueError:
            print("Invalid number, try again.")
            continue
        
        # Convert everything to meters
        if unit == 'm':
            figure = value
        elif unit == 'cm':
            figure = value * 0.01
        elif unit == 'dm':
            figure = value * 0.1
        elif unit == 'km':
            figure = value * 1000
        else:
            print("Unknown unit, try again.")
            continue
        
        print(f"\nConverted values:")
        print(f"Meters: {figure}")
        print(f"Centimeters: {figure * 100}")
        print(f"Decimeters: {figure * 10}")
        print(f"Kilometers: {figure / 1000}")
```
